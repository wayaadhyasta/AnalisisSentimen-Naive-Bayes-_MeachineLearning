```python
# mengubah text menjadi lower
data['text_cleaning'] = data['text_cleaning'].str.lower()
data.head()
```

```python
# normaliasi
# Membuat kamus normalisasi untuk mengganti kata tidak baku dengan kata baku
normalizad_word = {'dgn' : 'dengan',
                   'gue': 'saya','ngaco':'sembarangan',
                   'kwalitas':'kualitas',
                   'blom':'belum',
                   'aing':'aku',
                   'tehnologi':'teknologi'}
# Fungsi untuk melakukan normalisasi teks
def normalisasi(st_text):
  for i in normalizad_word: # Loop melalui setiap kata dalam kamus normalisasi
    str_text = str_text.replace(i,normalizad_word[i]) # Ganti kata tidak baku dengan kata baku
  return str_text # Kembalikan teks yang sudah dinormalisasi

  # Menerapkan fungsi normalisasi pada kolom 'text_cleaning' dalam DataFrame
  data['text_cleaning'] = data['text_cleaning'].apply(lambda x: normalisasi(x))
  data.head()
```

```python
#import Sastrawi
#from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
#more_stop_word = []

#stop_words = StopWordRemover().get_stop_words()
#new_array = ArrayDictionary(stop_words)
#stop_words_remover = StopWordRemover(new_array)

#def stopword(str_text):
 # str_text = stop_words_remover.remover(str_text)
 # return str_text

#data['text_cleaning'] = data['text_cleaning'].apply(lambda x: stopword(x))
#data.head()

# Stopword
import Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary

# Create a StopWordRemoverFactory object
stop_factory = StopWordRemoverFactory()

# Membuat objek StopWordRemoverFactory
stop_word_dictionary = stop_factory.create_stop_word_remover().dictionary

# Mendapatkan daftar stopword bawaan dari Sastrawi
stop_words_remover = StopWordRemover(stop_word_dictionary)

# Fungsi untuk menghapus stopword dari teks
def stopword(str_text):
  str_text = stop_words_remover.remove(str_text)  # Use remove() instead of remover()
  # Menggunakan remove() untuk menghapus stopword
  return str_text

# Menerapkan fungsi stopword removal pada kolom 'text_cleaning' dalam DataFrame
data['text_cleaning'] = data['text_cleaning'].apply(lambda x: stopword(x))
data.head()
```

```python
# tokenisasi
tokenized = data['text_cleaning'].apply(lambda x:x.split())
tokenized
```

```python
# Stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text_cleaning):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for i in text_cleaning:
    dt = stemmer.stem(i)
    do.append(dt)
  d_clean = []
  d_clean = ' '.join(do)
  print(d_clean)
  return d_clean

# lanjut dengan menerapakan variabel tokenize di dalam metode Stemming
tokenized = tokenized.apply(stemming)
tokenized.to_csv('/content/drive/MyDrive/mobil_listrik_bersih1.csv',index=False)
data_clean = pd.read_csv('/content/drive/MyDrive/mobil_listrik_bersih1.csv', encoding='latin1')
```

Data Preparation

```python
x = data_clean['text_cleaning']
y = data_clean['sentimen']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)
```

```python
x_train = x_train.astype(str)
x_test = x_test.astype(str)
# menggunakan CountVectorizer untuk mengubah teks menjadi fitur numerik
vectorizer = CountVectorizer()
x_train_vec = vectorizer.fit_transform(x_train.astype(str))  # Convert to string explicitly
x_test_vec = vectorizer.transform(x_test.astype(str))     # Convert to string explicitly
```

```python
# melakukan oversampling menggunakan SMOTE pada data latih
smote = SMOTE(random_state=42)
x_train_renampled, y_train_renampled = smote.fit_resample(x_train_vec, y_train)
```

```python
# contoh data hasil ranamepled
sentimen_counts = y_train_renampled.value_counts()

plt.figure(figsize=(5, 3))
plt.bar(sentimen_counts.index, sentimen_counts.values, color={'red','blue'})
plt.xlabel('Sentimen')
plt.ylabel('Jumlah')
plt.title('Jumlah Sentimen Setelah Oversampling')
plt.xticks(sentimen_counts.index, ['0','1'])
plt.show()
```

 Modeling

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(x_train_renampled, y_train_renampled)
```

```python
y_pred = naive_bayes.predict(x_test_vec)
```

```python
# Evaluasi Model
accuracy = accuracy_score(y_test, y_pred)
classification_report = classification_report(y_test, y_pred, target_names=['negatif','positif'])

print('Akurasi :',accuracy)
print("Laporan Klasifikasi:",classification_report)
```

Testing hasil

```python
new_text = input("\nMasukan teks baru untuk di coba: ")
new_text_vec = vectorizer.transform({new_text})
predicted_sentimen = naive_bayes.predict(new_text_vec)

if predicted_sentimen[0] == 1:
  # print("Sentimen Positif") tidak bisa pakai print karna hasil dari seleksi
  # tidak akan muncul informasi yg di input
  sentimen_label = 'Positif'
elif predicted_sentimen[0] == 0:
  # print("Sentimen Negatif") tidak bisa pakai print karna hasil dari seleksi
  # tidak akan muncul informasi yg di input
  sentimen_label = 'Negatif'
else:
  print("Sentimen Tidak Diketahui")

print("Hasil Analisis sentimen untuk teks baru :", sentimen_label)
```
